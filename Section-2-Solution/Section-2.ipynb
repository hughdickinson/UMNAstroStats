## Preliminaries
In order to complete the excercises in this section you will need to know how to generate arrays of random numbers that are sampled from a particular distribution.

To do this you can use the [`random`](https://docs.scipy.org/doc/numpy-1.15.1/reference/routines.random.html) module of the [`numpy`](https://docs.scipy.org/doc/numpy-1.15.1/) package.

````python
# Import the required modules and define aliases for the numpy package and the numpy.random module.
import numpy as np
import numpy.random as nprandom
````

### Uniformly distributed integers
To generate a **1D** array containing uniformly distributed **integers**, you can use the [`numpy.random.randint` function](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.randint.html#numpy.random.randint). The following example generates 100 integers that are uniformly distributed between 5 and 15.

````python
# Generate an array of 100 integers that are randomly distributed between 5 and 15
randomIntegerArray = nprandom.randint(low=5, high=15, size=100)

# Print the generated values
print("Array of 100 integers that are randomly distributed between 5 and 35:", 
      randomIntegerArray, 
      sep='\n')
````

### Uniformly distributed real-valued numbers
To generate a **1D** array containing uniformly distributed **real numbers**, you can use the `numpy.random.uniform` function.The following example generates 100 real numbers that are randomly distributed between 0.5 and 3.5.

````python
# Generate an array of 1000 floating point numbers that are randomly distributed between 0.5 and 3.5
randomFloatArray = nprandom.uniform(low=0.5, high=3.5, size=1000)

# Print every tenth generated value
print("100 floating point numbers that are randomly distributed between 0.5 and 3.5:", 
      randomFloatArray[::10], 
      sep='\n')
````

### Gaussian distributed real-valued numbers
To generate a **1D** array containing **Gaussian** distributed **real numbers**, you can use the [`numpy.random.normal`](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.normal.html#numpy.random.normal) function.The following example generates 1000 real numbers that are randomly drawn from a Gaussian distribution with mean (`loc`) 0.5 and **standard deviation** ($\sigma$, `scale`) 0.1.

````python
# Generate an array of 1000 floating point numbers that are are randomly drawn from a Gaussian distribution with 
# mean 0.5 and standard deviation 0.1
gaussianFloatArray = nprandom.normal(loc=0.5, scale=0.1, size=1000)

# Print every tenth generated value
print("100 floating point numbers drawn from Gauss(0.5, 0.1):", 
      gaussianFloatArray[::10], 
      sep='\n')
````

### Model Solution
#### Part 1
The stages of the _Monty Hall_ game can be strightforwardly modeled by drawing two random integers. We'll write code that implements the follwing algorithm.
 1. First model random placement of the car by drawing a random integer between 0 and 2.
 2. Second, model the player's initial choice by drawing a second random integer between 0 and 2.
 3. With these two values we can simulate the game's logic. Note that the outcome does not depend on Monty's choices or actions. In total, there are four possible pathways through the game.  
  * If the player's choice matches the car's position and they swap, they get a goat.
  * If the player's choice matches the car's position and they stick, they get the car.
  * If the player's choice matches a goat's position and they swap, they get the car.
  * If the player's choice matches a goat's position and they stick, they get a goat. 

````python
# Import the required packages and modules
import numpy as np
import numpy.random as nprandom

import matplotlib
import matplotlib.pyplot as mplplot

%matplotlib inline
````

The following cell contains two functions. The first (`playMontyHallOnce`) simulates a **single round** of the Monty Hall game and has comments that explain the program logic. The second function (`playMontyHallRepeatedly`) expects a numpy array of boolean values corresponding to player choices, and uses a very compact `numpy` boolean expression syntax.

````python
def playMontyHallOnce(playerSwaps):
    # Randomly pick the position of the car
    carPosition = nprandom.randint(3)
    # Randomly pick the player's choice
    playerChoice = nprandom.randint(3)
    # If the player picked the position of the car, Monty can show them 
    # one of two goats, but that doesn't matter, because...
    if playerChoice == carPosition:
        # ...if they swap...
        if playerSwaps:
            # ...then they "win" a goat...
            return False
        # ...but if they stick...
        else:
            # ...then they win the car!
            return True
    # On the other hand, if they picked a goat, Monty 
    # must show them the other goat...
    else:
        # ...so if they swap...
        if playerSwaps:
            # ...then they win the car...
            return True
        # ...but if they stick...
        else:
            # ...Goaty McGoatface!
            return False
        
def playMontyHallRepeatedly(playerSwaps):
    try:
        # Randomly pick the position of the cars in each round
        carPosition = nprandom.randint(3, size=playerSwaps.size)
        # Randomly pick the player's choices in each round
        playerChoice = nprandom.randint(3, size=playerSwaps.size)
        # Compact version of the game logic
        return ((playerChoice == carPosition) & ~playerSwaps) | ((playerChoice != carPosition) & playerSwaps)
    except AttributeError as e:
        print('An AttributeError was raised! Make sure that playerSwaps has a "size" attribute.\n Details: {}'.format(e))
        return None
````

When simulating large numbers of rounds, the second version runs much faster than invoking the first version multiple times. We can use the Jupyter [magic `%%timeit` functionality](https://ipython.readthedocs.io/en/stable/interactive/magics.html#magic-timeit) to demonstrate this.

````python
numRounds = 10000
````

First, repeatedly calling the single round version, assuming that the player always sticks with their initial choice.

````python
%%timeit -n 100
playerSwaps = False
results = []
for roundNumber in range(numRounds):
    results.append(playMontyHallOnce(playerSwaps))
````

Now, using the faster multi-round version.

````python
%%timeit -n 100
playerSwaps = ~np.ones(numRounds).astype(bool)
results = playMontyHallRepeatedly(playerSwaps)
````

On my laptop, that's over 100 times faster!

#### Part 2
The outcome of the game is purely random, so simulating 100 players (who all choose to stick with their original choice) playing 1000 rounds is exactly equivalent to 1 player playing 100,000 rounds.

Rather than making an array with 100,000 entries that are all `False`, we'll use the [`np.broadcast_to` function](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.broadcast_to.html), which creates a _view_ of a single `False` value that has all the properties of an array but only needs the memory required to store one boolean value. 

````python
numPlayers = 100
numRounds = 1000
results1D = playMontyHallRepeatedly(np.broadcast_to(False, numPlayers*numRounds))
````

Now we need to make a histogram of the fraction of times that each player won the car by sticking with their original choice. First let's manipulate our simulation output by changing its shape from a 1D array with 100,000 entries to a 2D array with 100 rows (representing the players) and 1000 columns (representing the games they played). We can use the [`numpy.reshape` function](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.reshape.html) to do this. We can use the `imshow function` from `matplotlib` to visualize the outcomes of each player's rounds, with a black cell for success and a white cell for failure.

````python
results2D = np.reshape(results1D, newshape=(numPlayers, numRounds))
# Plot the array as an image using the imshow function from matplotlib
results2DPlot = mplplot.figure(figsize=(10,3)).add_subplot(1,1,1)
results2DImage = results2DPlot.imshow(results2D, cmap='Greys', aspect='auto')
results2DPlot.set_xlabel('Round Number', fontsize='x-large')
results2DPlot.set_ylabel('Player Number', fontsize='x-large')
mplplot.tight_layout()
````

It's obviously not easy to work out which outcome is more likely for each player or in general. The next step towards answering the question is to compute the fraction of times each player wins the car. We can compute the **number** of times the indivdual players win the car by summing our results array over rounds. we can do this using the [`numpy.sum` function](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.sum.html) and specifying that we want to sum along the _Round Number_ axis of our array. Note that axis numbers are zero indexed, so we specify `axis=1` to select the _Round Number_ axis. Once we have the number of successed for all players we can simply divide by the number of rounds to compute the fraction of successes for each player.

````python
# Note that we explicitly cast numRounds to a float to avoid integer division
successFractions = np.sum(results2D, axis=1)/float(numRounds)
````

Finally, it's straightforward to make a histogram of the successFractions using the [`hist` function](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib.pyplot.hist).

````python
# Instantiate a figure to plot the histogram
successFractionFig = mplplot.figure(figsize = (8,8))
successFractionAxes = successFractionFig.add_subplot(1,1,1)
successFractionHist = successFractionAxes.hist(successFractions, bins=15, 
                                               label='{} rounds'.format(numRounds))
successFractionAxes.set_title('Success fraction distribution for {} players'.format(numPlayers), 
                              fontsize='x-large')
successFractionAxes.set_xlabel('Success fraction', fontsize='x-large')
successFractionAxes.set_ylabel('Number of players', fontsize='x-large')
successFractionAxes.legend(fontsize='x-large')
mplplot.tight_layout()
````

````python
def simulateMontyHall(playersSwap, numPlayers, numRoundsList):
    # Instantiate a figure to plot the histograms
    successFractionFig = mplplot.figure(figsize = (8,8))
    successFractionAxes = successFractionFig.add_subplot(1,1,1)
    
    # Define an array of bin boundaries so that all histograms are binned in the same way
    histBins = np.linspace(0, 1, 125)
    
    # Keep a record of the minimum and maxiumum values to be plotted
    minima = []
    maxima = []
    
    # Run the simulation and plot results assuming different numbers of rounds.
    for numRounds in numRoundsList:
        results1D = playMontyHallRepeatedly(np.broadcast_to(playersSwap, numPlayers*int(numRounds)))
        results2D = np.reshape(results1D, newshape=(numPlayers, int(numRounds)))
        successFractions = np.sum(results2D, axis=1)/float(numRounds)
        successFractionHist = successFractionAxes.hist(successFractions, bins=histBins, 
                                                       histtype='step', lw=2,
                                                       label='{:d} rounds'.format(int(numRounds)))
        
        # Record of the minimum and maxiumum success fractions for the current number of rounds
        minima.append(successFractions.min())
        maxima.append(successFractions.max())
        
    successFractionAxes.set_title('Success fraction distribution for {} players when {}.'.format(numPlayers, 'switching' if playersSwap else 'sticking'), 
                                  fontsize='x-large')
    successFractionAxes.set_xlabel('Success fraction', fontsize='x-large')
    successFractionAxes.set_ylabel('Number of players', fontsize='x-large')
    
    # Zoom the success fraction axis to encompass only the range of the data
    successFractionAxes.set_xlim(np.amin(minima), np.amax(maxima))
    
    successFractionAxes.legend(fontsize='x-large')
    mplplot.tight_layout()

simulateMontyHall(playersSwap=False, numPlayers=numPlayers, numRoundsList=[1e2, 1e3, 1e4, 1e5])
````

### Exercise 2 (AST5031 only):

Consider a bag in which there are red and blue balls, for a total of $N = 10$ balls.
Of the first three balls drawn from the bag, 2 are red and 1 is blue. Each ball is replaced immediately after it is drawn.

Use the Bayes’ and total probability theorems to compute the probability of having i red balls in the bag, given that we performed the measurement $B = \{$Two red balls were drawn within the first three trials$\}$. [Hint: you will need the Binomial distribution].

Write a program that demonstrate your answer is correct in the case of $i = 5$.

#### Part 1
This is simply an excercise in algebra, and a perfect chance to flex your Markdown muscles! First let's define:
* $n_{r}$ is the true number of red balls in the bag.
* $n_{b} = N-n_{r}$ is the true number of blue balls in the bag.

Now, since our measurement ($B$) yielded a sample containing one blue ball, we know that the set $\Omega = \{1\ldots9\}$ contains $n_{r}$. 

We're told to solve the problem using the total probability theorem
$$
P(B) = \displaystyle\sum\limits_{j}{P(B|A_{j})P(A_{j})}.
$$
as well as  Bayes's theorem 
$$
P(A|B) = \frac{P(B|A)P(A)}{P(B)} = \frac{P(B|A)P(A)}{\displaystyle\sum\limits_{j}{P(B|A_{j})P(A_{j})}}
$$

We're also told that
> Each ball is replaced immediately after it is drawn.

This is related to the hint we're given about the [Binomial distribution](https://en.wikipedia.org/wiki/Binomial_distribution). Since the balls are replaced immediately, we can treat each draw as a [Bernoulli Trial](https://en.wikipedia.org/wiki/Bernoulli_trial). In general, for a sequence of $n_{t}$ Bernoulli trials, each of which has a probability $p$ of success, the probability of observing $k$ successes can be computed using the [Binomial distribution](https://en.wikipedia.org/wiki/Binomial_distribution). 
$$
P(k|p) = \binom{n_{t}}{k}\;p^{k}(1-p)^{n_{t}-k}
$$

Definining $A$ to be the event that the bag contains $n_{r}=i$ red balls, then the probability that a single draw yields a red ball is:
$$
p(r|A) = 1-p(b|A) = \frac{i}{N}
$$
Using the binomial theorem, the probability of observing two red balls in a sequence of three draws (with immediate replacement), for arbitrary $i$ is:
$$
P(B|A) = \binom{3}{2}\;p(r|A)^{2}(1-p(r|A))^{3-2} = 3\cdot\left(\frac{i}{N}\right)^{2}\left(1-\frac{i}{N}\right)
$$
and since we don't know the true value of $i$, then all values are equally likely and $P(A)=1/N=0.1$,  
so Bayes's theorem becomes:
$$
P(A|B) = \frac{P(B|A)P(A)}
{\displaystyle\sum\limits_{j=0}^{N} {P(B|A_{j})P(A_{j})} } 
=\frac{3\cdot\left(\frac{i}{N}\right)^{2}\left(1-\frac{i}{N}\right)\cdot\frac{1}{N}}
{\displaystyle\sum\limits_{j=0}^{N}{3\cdot\left(\frac{j}{N}\right)^{2}\left(1-\frac{j}{N}\right)
\cdot\frac{1}{N}}} =\frac{\left(\frac{i}{N}\right)^{2}\left(1-\frac{i}{N}\right)}
{\displaystyle\sum\limits_{j=0}^{N}{\left(\frac{j}{N}\right)^{2}\left(1-\frac{j}{N}\right)
}} 
$$


#### Part 2
We need to write a program that simulates drawing three balls (with immediate replacement) from a bag of 10 balls of which $i=5$ are red. First, let's define a function to compute our analytic solution so we can check it against the simulation results.

````python
def redBallCountProbability(trueNumRed = 5, totalNumBalls = 10):
    binomialFactor = lambda numRed : (1-(numRed/totalNumBalls))*(numRed/totalNumBalls)**2
    numerator = binomialFactor(trueNumRed)
    denominator = np.sum([binomialFactor(possibleNumRed+1) for possibleNumRed in range(totalNumBalls)])
    return numerator/denominator
````

````python
analyticResult = redBallCountProbability(trueNumRed = 5, totalNumBalls = 10)
print('Analytic result:', analyticResult)
````

The process of drawing from the bag can be simulated by drawing a random, real valued number $x\in[0,1]$. For **any** value of $i$, we label the ball red if $x<\frac{i}{N}$ and blue otherwise. Now we can simulate a large number of draws for every possible value of $i\in{1\ldots10}$ and group them into disjoint sets of 3 to simulate many independent experiments. To empirically estimate $P(A|B)$ we need to count the fraction of experiments that yielded two red balls and one blue ball that **also** had $i=5$ red balls in the bag. 

````python
def simulateThreeBallDraws(trueNumRed = 5, totalNumBalls = 10, experimentsPerTrueNumRed = 1000000):
    twoRedBallCounts = []
    for possibleNumRed in range(1, totalNumBalls+1):
        # Generate as many random values in [0,1] as we require
        draws = nprandom.random(size=3*experimentsPerTrueNumRed)
        # Convert them to red/blue labels according to whether they exceed the i/N threshold
        ballIsRed = draws < float(possibleNumRed)/float(totalNumBalls)
        # Change the layout of the data so that the labeled draws are grouped into three-draw experiments
        experiments = ballIsRed.reshape((experimentsPerTrueNumRed, -1))
        # Identify experiments that have two red balls. 
        # Note that this is equivalent to exactly one blue ball.
        twoBallsAreRed = np.sum(~experiments, axis=1) == 1
        # Count the number of such experiments
        twoRedBallCounts.append(np.sum(twoBallsAreRed))
        
    # Now compute the fraction of experiments that had two red balls 
    # that also had trueNumRed balls in the bag.
    return twoRedBallCounts[trueNumRed-1]/float(np.sum(twoRedBallCounts))
````

````python
simulationResult = simulateThreeBallDraws(trueNumRed = 5, totalNumBalls = 10)
print('Simulation result:', simulationResult)
print('Simulation result - Analytic result:', simulationResult-analyticResult)
````

**Looks pretty close to me!** As a further test of whether our simulation matches the analytic expression, let's plot the results of both computations for all possible numbers of red balls. We'll mark the case when $i=5$ as the question requests.

Start by generating probabilities for each possible value of $i$.

````python
possibleRedBallCounts = list(range(1, 11))
simulatedProbs = [simulateThreeBallDraws(trueNumRed=numRed) for numRed in possibleRedBallCounts]
analyticProbs = [redBallCountProbability(trueNumRed=numRed) for numRed in possibleRedBallCounts]
````

Now plot the computed values.

````python
redBallCountAxis = mplplot.figure(figsize=(10,10)).add_subplot(1,1,1)
redBallCountAxis.scatter(range(1, 11), analyticProbs, marker='p', s=250, label='Analytic Values', zorder=5)
redBallCountAxis.scatter(range(1, 11), simulatedProbs, marker='*', s=50, label='Simulated Values', zorder=10) 
redBallCountAxis.set_xlabel('$N_{red}^{true}$', fontsize='x-large')
redBallCountAxis.set_ylabel('$P(i=N_{red}^{true}|B)$', fontsize='x-large')
redBallCountAxis.axvline(x=5, ls='--', c='k', lw=1, zorder=0,
                         label='$N_{{red}}^{{true}}=5,\;P(i=N_{{red}}^{{true}}|B)={:.3f}$'.format(simulatedProbs[4]))
redBallCountAxis.axhline(y=simulatedProbs[4], ls='--', c='k', lw=1, zorder=0)
redBallCountAxis.legend(fontsize='x-large', loc='upper left')
````

Finally, we should really verify that the total probability summed over all possible values of $i$ is equal to unity.

````python
analyticProbSum = np.sum(analyticProbs)
simulatedProbSum = np.sum(simulatedProbs)

print('Sum of analytic probabilities:', analyticProbSum)
print('Sum of simulated probabilities:', simulatedProbSum)
````
